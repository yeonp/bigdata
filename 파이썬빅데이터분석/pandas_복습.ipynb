```python
import pandas as pd
df = pd.read_csv('boston_train.csv')
df
```

```python
# 1.1 속성출력 : shape, index,value
print(df.shape)
print(df.index)
print(df.values)
```

```python
#  1.2 행(row) 으로 접근
df[:1]
df[11:16]
df[::3]
df[::-1]  # 행을 역순으로 출력
```

```python
# 1.3 열(columns)으로 접근
len(df.columns)  # 10
df.columns  # 컬럼 목록을 출력
df['CRIM']
df.CRIM
df[df.columns[0]]  # 세가지가 모두 동일한 결과
```

```python
df[df.columns[9]]

df.columns[-1]     # 'MEDV'
df[df.columns[-1]] # 마지막 컬럼( 9번 ,Y:답)
```

```python
# 1.4 열의 이름을 사용하여 열 추출(Series 객체 생성)
sr = df['CRIM']
sr
type(sr) # pandas.core.series.Series
sr.shape # (400,) 1차원 배열 
```

```python
# 1.5 열의 번호를 사용하여 열 추출(Series 객체 생성)
sr = df[df.columns[0]]
sr
type(sr) # pandas.core.series.Series
sr.shape # (400,) 1차원 배열 
```

```python
# 1.6 열의 이름을 사용하여 여러개 열을 추출
# 데이터 프레임 객체 생성
df2 = df[['CRIM','ZN','INDUS']]
df2
type(df2) # pandas.core.frame.DataFrame
```

```python
# 1.7 열의 번호를 사용하여 여러개 열을 추출
# 데이터 프레임 객체 생성
df3 = df[[df.columns[0],df.columns[1],df.columns[2]]]
df3
type(df3) # pandas.core.frame.DataFrame
```

```python
df4 = df[[df.columns[0],df.columns[1],df.columns[2]]][:5]
df4
```

```python
# 1.8 df.iloc[행번호,열번호]를 사용하여 원소 접근 및 수정
df.loc[5,'CRIM']
df.loc[:5,'CRIM']
df.loc[:5,['CRIM','ZN','INDUS']]
```

```python
df.iloc[0,0]
df.iloc[2,[0,1,2]]
df.iloc[:3,[0,1,2]]
```

```python
df.iloc[:3,[0,1,2]] = None  # 수정도 가능
df.iloc[:3,:3]
df
```

```python
# 1.9 수정된 데이터프레임을 새로운 CSV파일로 저장
df.to_csv('boston_train_new.csv',index=False)
```
